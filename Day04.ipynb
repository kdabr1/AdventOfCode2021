```python
import pandas as pd
df4a = pd.read_csv('4.txt', header=None, dtype = str)
df4a.head()
```

```python
#list_r - list of drawn numbers
list_r = df4a.iloc[0]
#.astype(int)
```

```python
list_r
```

```python
#df4a_2 - dataframe storing each table as a line
df4a_2 = pd.DataFrame(columns = range(25), index = range(100))
for a in range(int(df4a.iloc[1:,0].shape[0]/5)):
    for b in range (0,5):
        for c in range (0,5):
            df4a_2.iloc[a,5*b+c] = df4a.iloc[1:,0][5*a+b+1].split()[c]
df4a_2.head()
```

```python
#df4a_3 - dataframe where hits will be stored
df4a_3 = df4a_2.copy()
for i in range(len(list_r)): #drawing new number
    df4a_3.replace(list_r[i], True, inplace = True)
    for j in range(df4a_2.shape[0]): #checking each board      
        for a in range (5): #checking rows
            if df4a_3.iloc[j,5*a] == df4a_3.iloc[j,5*a+1] == df4a_3.iloc[j,5*a+2] == df4a_3.iloc[j,5*a+3] == df4a_3.iloc[j,5*a+4] \
            or df4a_3.iloc[j,a] == df4a_3.iloc[j,a+5] == df4a_3.iloc[j,a+10] == df4a_3.iloc[j,a+15] == df4a_3.iloc[j,a+20]:
                break                 
        else:
            continue
        break 
    else:
        continue
    break
    
print("Answer: ", df4a_3.iloc[j][df4a_3.iloc[j] != True].astype(int).sum()*int(list_r[i]))
```

```python
#df4a_4 - dataframe where hits will be stored
#boards - list of boards that already won
df4a_4 = df4a_2.copy()
boards = []
for i in range(len(list_r)): #drawing new number
    df4a_4.replace(list_r[i], True, inplace = True)
    for j in range(df4a_2.shape[0]): #checking each board      
        for a in range (5): #checking rows
            if df4a_4.iloc[j,5*a] == df4a_4.iloc[j,5*a+1] == df4a_4.iloc[j,5*a+2] == df4a_4.iloc[j,5*a+3] == df4a_4.iloc[j,5*a+4] \
            or df4a_4.iloc[j,a] == df4a_4.iloc[j,a+5] == df4a_4.iloc[j,a+10] == df4a_4.iloc[j,a+15] == df4a_4.iloc[j,a+20]:
                boards.append(j) if j not in boards else boards
            if len(boards) == 100:
                break
        else:
            continue
        break 
    else:
        continue
    break
    
print("Answer: ", df4a_4.iloc[boards[-1]][df4a_4.iloc[boards[-1]] != True].astype(int).sum()*int(list_r[i]))
```
