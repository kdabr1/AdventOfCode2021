```python
import pandas as pd
import itertools

df1 = pd.read_csv('9.txt', names = ['Input'], header=None)
df1.head()
```

```python
for i in range(len(str(df1['Input'].max()))):
    df1[i] = df1['Input'].astype(str).str[i]
df1.drop(labels = 'Input', axis = 1, inplace = True)
df1 = df1.astype(int)
df1.head()
```

```python
#Adding dummy rows and columns on the edges to avoid errors while checking the edges of the area.
df1.loc[-1] = 9  # adding a row
df1.index = df1.index + 1  # shifting index
df1 = df1.sort_index()  # sorting by index
df1.loc[max(df1.index) +1 ] = 9
df1.columns = df1.columns + 1
df1[0] = 9
df1[max(df1.columns) + 1] = 9
df1 = df1.reindex(sorted(df1.columns), axis=1)
df1
```

```python
x = 0
for col in range(1, max(df1.columns)):
    for row in range(1, max(df1.index)):  
        if df1[col][row] < min(df1[col-1][row], df1[col+1][row], df1[col][row-1], df1[col][row+1]):
                x += df1[col][row] + 1
print('Answer: ', x)
```

```python
#low_points - list of low points that will be used as a starting point for basins' size check
low_points = []
for col in range(1, max(df1.columns)):
    for row in range(1, max(df1.index)):  
        if df1[col][row] < min(df1[col-1][row], df1[col+1][row], df1[col][row-1], df1[col][row+1]):
                low_points.append([col, row])
```

```python
#basin_sizes - list storing sizes of all already checked basins
#basin_curr - list of coordinates of the basin that is currently being examined
#basin_checked - list of coordinates that have been already checked if the adjacent locations are part of this basin
#x, y - temp variables to store horizontal and vertical position of the location
#hor, ver - offset for checking adjacent locations
#i - supporting variable

basin_sizes = []
for lp in low_points:    
    basin_curr = [lp]
    basin_checked = []
    i = 0
    while len(basin_checked) != len(basin_curr):
        x = basin_curr[i][0]
        y = basin_curr[i][1]
        for hor, ver in ([0,-1], [0, 1], [-1, 0], [1,0]):
            if df1[x+hor][y+ver] != 9 and [x+hor, y+ver] not in basin_curr:
                basin_curr.append([x+hor, y+ver])
        basin_checked.append([x,y])
        i += 1
    basin_sizes.append(len(basin_curr))
basin_sizes.sort()
print('Anwer: ', basin_sizes[-1]*basin_sizes[-2]*basin_sizes[-3])
```
